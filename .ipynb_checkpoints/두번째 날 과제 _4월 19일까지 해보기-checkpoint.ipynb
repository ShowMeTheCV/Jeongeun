### 1. 이미지의 크기 속성         

- 이미지 크기란, 필드나 변수에 할당될 이미지의 너비 * 이미지의 높이를 의미한다. 이 __이미지는 행렬(matrix)형태로 구성되어 있다.__   
- 많은 알고리즘에서 이미지 크기를 변경하는 메서드인 Image Pyramide나 Resize를 전처리 과정에서 적용한다.   
- __단순히 연산할 이미지의 크기를 바꾸는 과정도 딥러닝의 정확도와 연산량에 큰 영향을 미친다.__      

- openCV에서 변수나 필드에서 __이미 설정된 이미지의 크기로 다른 이미지를 불러올 경우 오류가 나곤 하는데, 이는 이미지 크기 속성 값이 "액자 역할" 을 하기 때문이다.__    
- 불러온 이미지의 크기와 액자의 크기가 다르다면 오류가 날 수 밖에 없다 ㅠㅠㅠ.. 


> 왜 새로운 이미지를 불러왔을 때 오류가 나죠?    

> - 이미지를 불러서 담아오도록 지정한 변수에서 지정한 이미지의 크기와, 불러들여 저장할 이미지의 크기가 서로 달랐기 때문에.
> - 이 때는 원본 이미지의 크기를 변경하는 함수를 적용한 다음 변수나 필드로 불러와 사용한다.

### 2. 정밀도    

- __정밀도__ 란 : __이미지가 얼마나 많은 색상을 표현할 수 있는가?__ 를 의미한다.       
- 정밀도가 높을 수록 많은 색상을 표현할 수 있어 데이터 폭이 넓어지고, 더 자연스럽게 표시된다. 데이터의 처리 결과도 더 정밀해진다.   

- 비트 : 색상의 표현 개수를 설정     
- 표현 : nBit, nBit의 의미는 $2^n$ 이므로, 4bit는 $2^4$인, 16가지 색상 표현이 가능하다.      

> 색상을 표현할 때 적어도 몇 비트가 필요한가요?        
> - 적어도 8비트!


> n.uint8이 표현할 수 있는 범위를 정수로 표현하면 얼마인가요?      '
> - unsigned이므로 부호가 정해져있지 않고, 0부터 시작하여 $2^8$까지이므로 0~255


### 3. 채널      

- 채널 : 그래픽스 이미지의 __색상 정보__ 를 갖고 있다.             
- 어떤 채널을 사용하고, 몇 개의 채널을 사용할 지 여부가 알고리즘과 연산량에 큰 영향을 미친다.

> 채널은 총 몇개인가요?       
> - 7개(RGB+알파+Hue, saturation, value..)

> 흑백 이미지를 가져올 때 왜 단일 채널을 사용해야 하나요? 3채널을 쓰면 안되나요?    
> 1채널로 밝기만 표현해줘도 된다. 그 외에 더 ++ 하게 되면 동일한 값이 3번 반복되며 쓸데없이 연산량을 늘릴 뿐이다.



### 4. 관심영역     


> 1920X1280 비율의 컬러 이미지를 가져온다면 몇 개의 데이터 공간이 생성되나요?          
> 1920*1280*3 = 7,382,800개

- __관심영역(ROI, Region of Interest) : 이미지 상에서 관심있는 영역을 의미한다.__            
- 데이터가 커질 때 마다 기하급수적으로 처리할 데이터가 늘어나므로, 이를 방지하고자 관심 영역을 사용해 불필요한 데이터의 양을 줄인다.    

- __이미지를 처리할 때, 객체를 탐지하거나 검출하는 영역을 명확하게 관심 영역이라 지정할 수 있다.__  (ex)바운딩 박스)       


> 관심 영역을 사용해야 하는 이유를 설명한다면요?    
> - 관심영역을 설정하고, 그 영역을 제어함으로써 불필요한 데이터의 양을 줄인다. + 관심 영역에 대한 데이터만 온전하게 보존하여 __알고리즘 정확도와 연산 속도를 높인다.__ 

### 5. 관심 채널       

- __관심 채널(COI, Channel of Interest) : 이미지 상에서 관심있는 채널__       

> 관심 채널은 어디에 활용되나요?       
> - 모든 영역에서 데이터의 형태를 변형하는 것이 아니라, 특정 영역을 불러와 연산량을 감소시키고 정확도를 올리는 데 사용된다.


### 6. 히스토그램 


> 히스토그램이란 무엇인가요?       
> - 도수분포표, 데이터의 분포를 몇 개의 구간으로 나누고 각 구간에 속하는 밸류값을 시각적으로 표현한 막대그래프

> 히스토그램이 갖는 세 가지 중요 요소를 말하고, 이를 설명해주세요.       
> - 빈도수(X축 간격), 차원 수(히스토그램을 분석할 이미지의 차원), 범위(히스토그램의 x축 범위)

> 어떤 흑백 이미지를 가져와, 히스토그램을 출력해주세요.(예제 2.8 수행)        

```python
#차이 https://www.programmersought.com/article/85174804741/
import cv2
import numpy as np

import os
from matplotlib import pyplot as plt

image_path = os.getenv('HOME') + '/ShowMeTheCV/Jeongeun/PostingPic/1_histogram.jpg'
img_gray = cv2.imread(image_path)

image_path
```

```python
plt.imshow(img_gray)
plt.show()
```

```python
gray = cv2.cvtColor(img_gray, cv2.COLOR_BGR2GRAY)
result = np.zeros((img_gray.shape[0],256), dtype=np.uint8)

hist = cv2.calcHist([img_gray], [0], None, [256], [0,256])
cv2.normalize(hist, hist, 0, 255, cv2.NORM_MINMAX)

for x, y in enumerate(hist):
    cv2.line(result, (x, img_gray.shape[0]), (x, img_gray.shape[0]-y), 255)
    
dst = np.hstack([img_gray[:,:,0], result])
cv2.imshow("dst", dst)
```

### 2. ndarray 클래스(n차원 배열)                  

- __ndarray 클래스는 동일한 자료형을 가진 값들이 행렬 형태로 N차원 배열을 생성한다.__   

> np.array 함수에서 C스타일과 Fortran스타일의 메모리 레이아웃의 차이를 설명해주세요.      
> - C에서의 메모리 순서 : 가장 빠르게 변화하는 인덱스 순서로 할당된다.        
> - Fortran 메모리 순서 : 앞의 배열 색인이 최댓값에 도달해야 뒤로 넘어가게 된다.
> - (i)(j)(k) 순서일 때,         
> - C에서는 k부터 바뀐다. (0)(0)(0) -> (0)(0)(1)      
> - 포트란에서는 i부터 바뀐다. (0)(0)(0) -> (9)(0)(0) -> (9)(1)(0)

> 예제 3.23을 수행하여 출력값을 확인하세요.     

```python
import numpy as np

#값을 스스로 넣어주는 방법
array1 = np.array([[1,2,3],[4,5,6]])

#dtype=complex(복소수), 차원은 3으로 스스로 생성
array2 = np.array([1,2,3], dtype=complex, ndmin=3)

#array1을 카피=Flase
#여기서 copy=True 라면 array1의 값을 그대로 가짐
array3 = np.array(array1, copy=True)

#subok=True이면 matrix클래스, False이면 ndarray클래스
array4 = np.array(np.mat('12;34'), subok=True)


array1[0] = [4,5,6]

print(array1)
print(array2)
print(array3)
print(type(array4))
print(array4)-
```

#### 배열의 차원 변형   

> 예제 3.27을 수행하고 출력값을 확인하세요.     

```python
import numpy as np

#0에서 12까지
array = np.arange(12)

#2,3,2 의 크기로 리셰이프 [0,0][0,0][0,0] * 2
reshape1 = array.reshape(2,3,2)

#맨 처음만 2로 맞추고, 나머지는 알아서 맞춤(-1)
reshape2 = np.reshape(array, (2, -1), order='F')


print(type(array))

print(reshape1)
print(type(reshape1))

print(reshape2)
```

> - array.reshape와 np.reshape로 변형시킬 수 있음

```python
import numpy as np

array = np.arange(4)

axis1 = array[np.newaxis]
axis2 = array[:, np.newaxis]

print(array)
print(axis1)
print(axis2)
```

> 예제 3.29을 수행하고 출력값을 확인해주세요. 

```python
import numpy as np

array = np.arange(12).reshape(3, 4)

print(array)

flat1 = array.flatten(order='F')

#ravel : 풀다, 1차원으로 변형
flat2 = array.ravel()

print(flat1)
print(flat2)
```

#### 배열의 병합 분리 

> 예제 3.30을 수행하고 출력값을 확인해주세요. 

```python
import numpy as np

array1 = np.arange(6).reshape(2,3)
array2 = np.arange(6, 12).reshape(2,3)

merge1 = np.stack([array1, array2], axis=0)
merge2 = np.stack([array1, array2], axis=1)

#axis=0은 첫번째 차원을 기준으로
#axis=1은 두번째 차원을 기준으로
print("merge1")
print(merge1)

print("merge2")
print(merge2)
```

#### 배열 연산          

- 파이썬에 기본 내장된 math 라이브러리는 실수에 대해서만 연산을 지원한다.     
- 따라서, 컨테이너 자료형에 대해서는 연산이 불가해서 반복문으로 배열 연산을 수행했다.  

Numpy의 경우는 브로드 캐스팅 및 형식 캐스팅 기능을 ndarray 클래스로 지원하며, 배열의 요소별 연산을 고속으로 수행하는 벡터화된 래퍼 함수(wrapper function)이다.   

> 브로드캐스팅이란 무엇인가요?      
- numpy에서 차원의 크기가 서로 다른 배열에도 산술 연산을 가능하게끔 복제하여 적용하는 원리

- (1 20)       (1.5  x)                   (2.5 0+x)                 
  (0 0)   +   (x  x)               인 경우 (0+x 0+x) 로 수행함         

> 형식 캐스팅은 무엇인가요?       
- 연산하려는 두 배열의 자료형을 비교해 표현 범위가 더 넓은 자료형을 선택하는 것.
- int vs float일 경우 float으로 결과값을 얻는다.

### Matrix 클래스       

- 입력 배열을 행렬로 해석해서 사용하는 클래스이다!  
- __Matrix 클래스는 행렬 연산에 특화된 2차원 배열이다.__        

> 예제 3.34을 수행하고 출력값을 확인해주세요.        

```python
import numpy as np

array1 = np.array([1,2,3,4]).reshape(2,2)
array2 = np.array([5,6,7,8]).reshape(2,2)

mat1 = np.mat(array1)
mat2 = np.mat(array2)

print(mat1.T * mat2)
print(mat1 * mat2)

print(2 * array1)
```

- mat 클래스는 행렬 연산에 특화된 2차원 배열이다.   

> np.ndarray의 곱, 제곱 연산과 다른 점은 무엇인가요?      

- ndarray는 곱과 제곱 연산을 각 원소에 대해 수행한다.(아다마르 곱)
- mat 연산은 곱과 제곱을 행렬 간의 연산으로 처리한다.

### np.array으로 관심 영역 /관심 채널 설정하기      

> 예제 3.35을 수행하고 출력값을 확인해주세요.         

```python
import numpy as np

array = np.zeros((1280, 1920, 3), np.uint8)

x, y, w, h = 100, 100, 300, 300
roi = array[x:x+w, y:y+h]

print(array.shape)
print(roi.shape)
```

> 예제 3.36을 수행하고 출력값을 확인해주세요. 

```python
import numpy as np

array = np.zeros((1280, 1920, 3), np.uint8)
coi = array[:,:,0]

print(array.shape)
print(coi.shape)
```
